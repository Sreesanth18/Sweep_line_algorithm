```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Test Case 1
#plt.rcParams["figure.figsize"] = [7.5,3.5 ]
#plt.rcParams["figure.autolayout"] = True

plt.xlim( [ 0, 15] )          # Plot from x=0 to x=80.
plt.ylim( [ 0, 15 ] )         # Plot from y=0 to y=250.
plt.xticks( range(0,15,1) )   # Put x axis ticks every 10 units.
plt.yticks( range(0,15,1) )  # Y ticks every 50.  You can provide any list.
masterlist=[ [ [10,1],[10,1] ], [ [2,4],[4,2] ] , [ [4,6], [6,4] ] , [ [6,8], [8,6] ], [ [8,10],[10,8] ]]
for line in masterlist:
    plt.plot(line[0], line[1], 'bo', linestyle="-")
    plt.text(line[0][0]-0.015, line[1][0]+0.25, (line[0][0],line[1][0]))
    plt.text(line[0][1]-0.015, line[1][1]+0.25, (line[0][1],line[1][1]))

x = np.array([3,5,7,9])
y = np.array([3,5,7,9])

plt.scatter(x, y,color = 'black')
```
